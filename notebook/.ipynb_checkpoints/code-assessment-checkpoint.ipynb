````python

from sqlalchemy import create_engine
import pandas as pd
from IPython.display import Image
import psycopg2
from configparser import ConfigParser

config_path=r'/home/jovyan/config/'  # DB creds stored in a config file

def config(filename, section):       # function read config file and op params
    parser = ConfigParser()
    parser.read(filename)
    
    db = {}
    if parser.has_section(section):
        params = parser.items(section)
        for param in params:
            db[param[0]] = param[1]
    
    return db


def get_engine(schema='public'):          # function returns sql connection used by pandas
    dbschema=schema
    param=config(filename=config_path+r'database.ini', section='postgresql')
    db_uri='postgresql+psycopg2://'+param['user']+':'+param['password']+'@'+param['host']+'/'+param['database']
    engine=create_engine(db_uri,connect_args={'options': '-csearch_path={}'.format(dbschema)})
    return engine
````

### <font color='Orange'>Removing Duplicates

<font color='Red'> **NOTE : while working on ETL exercise after doing more exploratory analysis on data I feel like there are not any duplicates and one encounter can have multiple procedures performed on the patient. But in this exercise I remove duplicates based on encounter id.
    
<font color='black'> Lots of procedures data have same encounter id diffrent start and stop date but same time.
   
For simplicity and keeping only procedures file in scope, I will select the first record for a given encounter ID to remove duplicates for this exercise.
    
Before doing any operation on data I am removing duplicates so that other operations are comparatively faster.

````python
procedures.drop_duplicates(subset=['ENCOUNTER'],inplace=True) # Removing duplicate encouter IDs
````

### <font color='orange'>Procedure Duration in Seconds

````python
procedures['START']=pd.to_datetime(procedures['START']) # changing datatype to datetime
procedures['STOP']=pd.to_datetime(procedures['STOP'])
````

````python
procedures.info() # checking counts and datatypes  
````

````python
procedures['DURATION']=procedures['STOP']-procedures['START'] # calculating time diffrence
````

````python
procedures['DURATION']=procedures['DURATION'].apply(lambda x : x.seconds) # converting timedelta to seconds
````

````python
procedures.head()
````

### <font color='Orange'>Extracting Columns
    
<font color='black'> Generating new columns YEAR, MONTH, WEEK, DAY from START Timestamp

````python
procedures['YEAR']=procedures['START'].apply(lambda x : x.year)
procedures['MONTH']=procedures['START'].apply(lambda x : x.month_name())
procedures['WEEK']=procedures['START'].apply(lambda x : x.isocalendar().week)
procedures['DAY']=procedures['START'].apply(lambda x : x.day_name())
````

### <font color='orange'>Rank the payer by cost paid
    
#### <font color='green'>"payer_coverage" column is used to calculate cost paid by payer, considering company will pay only for covered amount not beyond that.

````python
final_df=pd.read_sql("""
select p.name,sum(payer_coverage)::integer "cost payed by company" from curated.fact_encounters fe
left outer join curated.dim_payer p on fe.payer=p.id
group by 1
order by 2 desc;
"""
,con=get_engine())                                                                 # read required data from database

final_df.to_csv(r'/home/jovyan/output/payer_by_cost.csv', index=False)  # store as csv file

final_df
````

### <font color='orange'>Top 5 highest costing patients (from payer prespective)
    
#### <font color='green'>"payer_coverage" column is used to calculate cost paid by payer for a patient.

````python
final_df=pd.read_sql("""
select pat.id,pat.first||' '||pat.last "patient name",p.name "company name",sum(payer_coverage) "cost payed by company" from curated.fact_encounters fe
left outer join curated.dim_payer p on fe.payer=p.id
left outer join curated.dim_patient pat on fe.patient=pat.id
group by 1,2,3
order by 4 desc limit 5;
"""
,con=get_engine())                                                                 # read required data from database

final_df.to_csv(r'/home/jovyan/output/payer_by_cost_company.csv', index=False)  # store as csv file

final_df
````

### <font color='orange'>Top 5 highest costing patients (from Patient prespective)
    
#### <font color='green'>"total_clain-cost" - "payer_coverage" column is used to calculate cost paid by patient from own pocket.

````python
final_df=pd.read_sql("""
with base as (
select pat.id,pat.first||' '||pat.last "patient name",p.name,sum(total_claim_cost) "cost_payed_total",sum(payer_coverage) "cost_payed_company" from curated.fact_encounters fe
left outer join curated.dim_payer p on fe.payer=p.id
left outer join curated.dim_patient pat on fe.patient=pat.id
group by 1,2,3 
)
select id,"patient name",sum(cost_payed_total) - sum(cost_payed_company) "cost payed by self"
from base
group by 1,2
order by 3 desc limit 5
"""
,con=get_engine())                                                                 # read required data from database

final_df.to_csv(r'/home/jovyan/output/payer_by_cost_self.csv', index=False)  # store as csv file

final_df
````

### <font color='orange'>top 5 most expensive procedures on daily basis
    
#### <font color='green'>first I select top 5 costly procedures daily based on "base_cost" (there can be more than 5 procedures if cost is same), then I count the unique days those procedures were performed and sort them on mostly ocurred.
    
this is not perfect because costly procedures can be less frequent, and also if some treatment is happening frequently will affect the results.
    

````python
final_df=pd.read_sql("""
with daily_top_5 as (
select t.description,DATE(start) "date" ,base_cost, rank() over (partition by DATE(start) order by base_cost desc) "rnk" from curated.fact_procedures fp
left outer join curated.dim_treatment t on fp.code=t.code
)
select description,count(distinct "date") freq
from daily_top_5
where rnk <=5
group by 1 order by 2 desc limit 5;
"""
,con=get_engine())                                                                 # read required data from database

final_df.to_csv(r'/home/jovyan/output/top_5_procedures.csv', index=False)  # store as csv file

final_df
````
